3)Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

```python

x = record1.aggregate(
[{
    "$project" :
    {
        "name":1,
        "_id" : 0,
        "result":{
            "$cond":
                    {"if": {"$and" : [{"$gte" : ["$exam_score", 40]}, {"$gte" : ["$quiz_score", 40]}, {"$gte" : ["$homework_score", 40]}]},
                    "then" :"pass",
                    "else":"fail"
                 }
                }
    }
}])
for i in x:
    print(i)
```
